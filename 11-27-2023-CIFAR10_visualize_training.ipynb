```python
from training import *
from analysis import *
from network import *

hidden_dim, n_synapse = 10, 10
model_type, loss_type = 'parallel', 'nll'
bias = 0

assert model_type in ['parallel', '2nn']
assert loss_type in ['nll', 'hinge']
# dataset = 'MNIST'
dataset = 'CIFAR10'

if dataset == 'MNIST':
    input_dim = 28*28
elif dataset == 'CIFAR10':
    input_dim = 32*32*3

hidden_range = (0, 10) if hidden_dim < 40 else (0, 20)
# bias = 10 if hidden_dim < 40 else 20
# bias = 0 if loss_type == 'hinge' else bias
    
if model_type == 'parallel':
    model = ParallelSynapse2NN(input_dim=input_dim,
        hidden_dim = hidden_dim, 
                            n_synapse = n_synapse,
                            hidden_range = hidden_range, 
                            additive_bias = bias)
    
elif model_type == '2nn':
    model = TwoLayerNN(input_dim=input_dim, hidden_dim = hidden_dim)


train_loader, test_loader = get_loader(dataset)
```

```python
plot_parallel_synapse_params(model)
plot_input_histogram_to_parallel_synapse_layer(model, test_loader, hidden_act='relu')
```

```python
for i in range(10):
    result = train_models(model, 
                        input_dim,
                    train_loader,
                    test_loader,
                    model_type=model_type,
                    loss_type=loss_type,
                    num_epochs=5)
    plot_parallel_synapse_params(result['model'])
    plot_input_histogram_to_parallel_synapse_layer(result['model'], test_loader, hidden_act='relu')

```
